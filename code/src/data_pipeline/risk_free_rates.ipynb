## Risk Free Rate Analysis

In this notebook, we plot the historical risk free rates in the US from 1990 to 2024 and calculate the average risk free rate from 2000 onward. 

The data is sourced from the Fama-French dataset, and is publicly available at `https://mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html`



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../../data/raw/us_risk_free_rates.csv')

df.rename(columns={df.columns[0]: 'Date'}, inplace=True)
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')

# Filter for the years 1990 to 2020
df_filtered = df[(df['Date'] >= '1990-01-01')]

plt.figure(figsize=(12, 6))
plt.plot(df_filtered['Date'], df_filtered['RF'])
plt.title('US Risk Free Rate (1990 - 2020)')
plt.xlabel('Date')
plt.ylabel('Risk Free Rate (RF)')
plt.grid(True)
plt.show()
```

```python
df_2000_onward = df[df['Date'] >= '2000-01-01']
average_rf = df_2000_onward['RF'].mean()
print(f"Average Risk-Free Rate (daily) from 2000 onward: {average_rf:.6f}")
```
